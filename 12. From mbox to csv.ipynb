```python
def mbox_to_csv(mbox_file, output_csv):
    with open(output_csv, 'w', newline='', encoding='utf-8') as csvfile:
        fieldnames = ['From', 'To', 'Date', 'Subject', 'Body']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

        writer.writeheader()

        mbox = mailbox.mbox(mbox_file)

        for message in mbox:
            from_ = message['from']
            to = message['to']
            date = message['date']
            subject = message['subject']
            body = get_body(message)

            writer.writerow({'From': from_, 'To': to, 'Date': date, 'Subject': subject, 'Body': body})
```

```python
def get_body(message):
    if message.is_multipart():
        for part in message.walk():
            if part.get_content_type() == 'text/plain':
                return part.get_payload(decode=True).decode('utf-8', errors='ignore')
    else:
        return message.get_payload(decode=True).decode('utf-8', errors='ignore')
    return ''
```

```python
if __name__ == "__main__":

    # Lista todos os ficheiros na pasta mbox_dir
    mbox_files = [f for f in os.listdir(folder_path) if f.endswith('.mbox')]

    for mbox_file in mbox_files:
        mbox_path = os.path.join(folder_path, mbox_file)
        output_csv_file = os.path.join(folder_path, mbox_file.replace('.mbox', '.csv'))

        mbox_to_csv(mbox_path, output_csv_file)
```
