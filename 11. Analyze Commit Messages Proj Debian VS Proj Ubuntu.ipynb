```python
def extract_user(email):
    match = re.search(r'^([^@]+)@', email)  # Match text before the @
    if match:
        user = match.group(1)  # Get the text before the @
        return user
    return None
```

# DPKG

```python
# Define the path to the CSV file for dpkg repository
dpkg_debian = folder_path + "dpkg_deb.csv"
dpkg_ubuntu = folder_path + "dpkg_ubu.csv"

# Read commit history from the CSV file
dpkg_deb_df = pd.read_csv(dpkg_debian)
dpkg_ubu_df = pd.read_csv(dpkg_ubuntu)
```

```python
dpkg_deb_df.head()
```

```python
dpkg_ubu_df.head()
```

### Comparison by Contributor Name (Full Match)

```python
# Extraindo os autores únicos de cada dataframe
authors_debian = set(dpkg_deb_df['author'].unique())
authors_ubuntu = set(dpkg_ubu_df['author'].unique())

# Encontrando os autores comuns
common_authors = authors_debian.intersection(authors_ubuntu)

# Criando o novo dataframe com os autores comuns
df_common_authors = pd.DataFrame({
    'contributor_debian': list(common_authors),
    'contributor_ubuntu': list(common_authors)
})

# Exibindo o novo dataframe
df_common_authors
```

### Comparison by Contributor Email (User ID)

```python
# Aplicando a função extract_user para extrair os nomes dos emails
dpkg_deb_df['user'] = dpkg_deb_df['email'].apply(extract_user)
dpkg_ubu_df['user'] = dpkg_ubu_df['email'].apply(extract_user)

# Extraindo os nomes únicos de cada dataframe
users_debian = set(dpkg_deb_df['user'].unique())
users_ubuntu = set(dpkg_ubu_df['user'].unique())

# Encontrando os nomes comuns
common_users = users_debian.intersection(users_ubuntu)

# Criando o novo dataframe com os nomes comuns
df_common_users = pd.DataFrame({
    'contributor_debian': list(common_users),
    'contributor_ubuntu': list(common_users)
})

# Exibindo o novo dataframe
df_common_users
```

# GLIBC

```python
# Define the path to the CSV file for dpkg repository
glibc_debian = folder_path + "glibc_deb.csv"
glibc_ubuntu = folder_path + "glibc_ubu.csv"

# Read commit history from the CSV file
glibc_deb_df = pd.read_csv(glibc_debian)
glibc_ubu_df = pd.read_csv(glibc_ubuntu)
```

```python
glibc_deb_df.head()
```

```python
glibc_ubu_df.head()
```

### Comparison by Contributor Name (Full Match)

```python
# Extraindo os autores únicos de cada dataframe
authors_debian = set(glibc_deb_df['author'].unique())
authors_ubuntu = set(glibc_ubu_df['author'].unique())

# Encontrando os autores comuns
common_authors = authors_debian.intersection(authors_ubuntu)

# Criando o novo dataframe com os autores comuns
df_common_authors = pd.DataFrame({
    'contributor_debian': list(common_authors),
    'contributor_ubuntu': list(common_authors)
})

# Exibindo o novo dataframe
df_common_authors
```

```python
# Aplicando a função extract_user para extrair os nomes dos emails
glibc_deb_df['user'] = glibc_deb_df['email'].apply(extract_user)
glibc_ubu_df['user'] = glibc_ubu_df['email'].apply(extract_user)

# Extraindo os nomes únicos de cada dataframe
users_debian = glibc_deb_df[['author', 'user']].drop_duplicates()
users_ubuntu = glibc_ubu_df[['author', 'user']].drop_duplicates()

# Encontrando os contribuidores comuns
common_authors = pd.merge(users_debian, users_ubuntu, on='author', suffixes=('_debian', '_ubuntu'))

# Renomeando as colunas conforme necessário
df_common_users = common_authors.rename(columns={
    'author': 'contributor',
    'user_debian': 'id_debian',
    'user_ubuntu': 'id_ubuntu'
})

# Selecionando e renomeando as colunas finais
df_common_users = df_common_users[['contributor', 'id_debian', 'contributor', 'id_ubuntu']]
df_common_users.columns = ['contributor_debian', 'id_debian', 'contributor_ubuntu', 'id_ubuntu']

# Exibindo o novo dataframe
df_common_users
```
